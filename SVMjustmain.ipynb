```python
import librosa, librosa.display
import numpy as np
import matplotlib.pyplot as plt
import os
import scipy
import math
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import r2_score
```

```python
def just_get_main(df):
    df_temp=df['mfcc_'+str(0)+'_avg']
    df_temp=df_temp.rename('blah')
    df_temp=pd.concat([df_temp, df['main_freq']], axis=1)
    df_temp=pd.concat([df_temp, df['max_mean']], axis=1)
    return df_temp.drop(columns=['blah'])

def SVC_rbf_on_maindf(df, C=10, crit=False):
    X=just_get_main(df=df)
    if crit:
        Y = df['critter_name']
    else:
        Y = df['fam_or_subfam']
    X_train, X_test, y_train, y_test = train_test_split(X.copy(), Y, shuffle=True, random_state=17, test_size=.2, stratify=Y)
    svc_pipe = Pipeline([('scale', StandardScaler()), ('svc_rbf',SVC(kernel='rbf', C=C))])
    svc_pipe.fit(X_train, y_train)
    return svc_pipe, X_train, X_test, y_train, y_test

def stats(model, X_set,y_set):
    predicted_vector = model.predict(X_set)
    true_vector = y_set
    conf_mat = confusion_matrix(true_vector, predicted_vector)
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    return conf_mat, accuracy

def convert_labels(labels, df, from_division, to_division):
    lookup_table = dict() # look up table for new label
    for label in labels:
        if label in lookup_table.keys():
            continue
        idx = df.loc[df[from_division]==label].index[0]
        lookup_table[label] = df[to_division][idx]
    return [lookup_table[label] for label in labels]
    
def stats_convert(model, df, X_set,y_set):
    predicted_vector = model.predict(X_set)
    true_vector = y_set
    true_class = convert_labels(true_vector, df, 'fam_or_subfam', 'critter_name')
    pred_class = convert_labels(predicted_vector, df, 'fam_or_subfam', 'critter_name')
    conf_mat = confusion_matrix(true_class, pred_class)
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    return conf_mat, accuracy
```

```python
df = pd.read_csv('MLNS_05292024.csv')
```

```python
#Runs svc rbf over the indicated C and depth of all 4 mfcc features, including Calvin's features and peak_freq
Ctests = [130]
for C in Ctests:
    print ('SVCRBF with C = ', C)
    model, X_train, X_test, y_train, y_test = SVC_rbf_on_maindf(df,C=C)
    conf_matrix, accuracy = stats(model, X_train,y_train)
    print('in training got confusion matrix')
    print(pd.DataFrame(conf_matrix))
    print ('with accuracy ', accuracy)
    print('in test got confusion matrix')
    conf_matrix, accuracy = stats(model, X_test,y_test)
    print(pd.DataFrame(conf_matrix))
    print ('with accuracy ', accuracy)
    print ('')
```

```python
conf_matrix, accuracy = stats_convert(model, df, X_train,y_train)
print('in training got confusion matrix')
print(pd.DataFrame(conf_matrix))
print ('with accuracy ', accuracy)
print('in test got confusion matrix')
conf_matrix, accuracy = stats_convert(model, df, X_test,y_test)
print(pd.DataFrame(conf_matrix))
print ('with accuracy ', accuracy)
print ('')
```
