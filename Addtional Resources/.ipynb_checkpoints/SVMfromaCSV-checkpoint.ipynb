```python
import os
import numpy as np
import pandas as pd
import math
import librosa
import scipy
import matplotlib.pyplot as plt
import time
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

Running with mfcc stats

.

```python
#store the dataframe with MFCC coefficients drop the grasshopper
df = pd.read_csv('MLNS_with_mfcc_stats_05242024.csv')
df=df.drop(33)

mfccs = []
for i in range(0,40):
    string1 = 'mfcc_' + str(i) + '_avg' 
    mfccs.append(string1)
    string1 = 'mfcc_' + str(i) + '_var'
    mfccs.append(string1)

cols = ['critter_name','fam_or_subfam'] + mfccs

data = df[cols]
data.head(5)
```

```python
short_label = data['fam_or_subfam'].values
le = LabelEncoder()
songs_label_num = le.fit_transform(short_label)


data_matrix = data[mfccs].values
data_label = songs_label_num


#scale the data
scaler = StandardScaler()
data_scale = scaler.fit_transform(data_matrix)

X_train, X_test, y_train, y_test = train_test_split(data_scale, data_label, 
                                                    test_size = .2, random_state = 404)

X_tt, X_val, y_tt, y_val = train_test_split(X_train, y_train, test_size = .2, random_state = 202)
print('Train validation split train shape: ', X_tt.shape)
```

```python
C = 1
accs = np.zeros((35,2))
for n in range(5,40):
    #print ('RBF kernel model with C = ', C, ' and applied to ', n ,' mfcc_coefficients')
    start = time.time()
    model = SVC(kernel='rbf', C=C, max_iter=500000)
    model.fit(X_tt[:,0:2*n], y_tt)
    end = time.time()
    print('time training: for C = ', C, ' was ' , end - start)
    
    #print('on test set')
    predicted_vector = model.predict(X_val[:,0:2*n])
    true_vector = y_val
    conf_mat = confusion_matrix(true_vector, predicted_vector)
    #print(pd.DataFrame(conf_mat))#,
    #                 columns = ['Predicted 0', 'Predicted 1', 'Predicted 2', 'Predicted 3'],
    #                 index = ['Actual 0', 'Actual 1', 'Actual 2', 'Actual 3'])
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    #print('accuracy =', accuracy)
    accs[n-5,0] = accuracy
    
    
    #print('on training set')
    predicted_vector = model.predict(X_tt[:,0:2*n])
    true_vector = y_tt
    conf_mat = confusion_matrix(true_vector, predicted_vector)
    #print(pd.DataFrame(conf_mat))#,
    #                 columns = ['Predicted 0', 'Predicted 1', 'Predicted 2', 'Predicted 3'],
    #                 index = ['Actual 0', 'Actual 1', 'Actual 2', 'Actual 3'])
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    #print('training accuracy =', accuracy)
    #print ('')
    accs[n-5,1] = accuracy
```

```python
print ('best test accuracy at n=', np.argmax(accs[:,0]) + 5, ' with accuracy ', accs[np.argmax(accs[:,0]),0])
print ('at n=15 test accuracy is already ', accs[10,0])
print ('best train accuracy at n=', np.argmax(accs[:,1]), ' with accuracy ', accs[np.argmax(accs[:,1]),1])
## Make a figure
plt.figure(figsize = (7,7))

plt.plot(accs[:,0])
plt.plot(accs[:,1])

for i in range(0,35):
    plt.scatter( i ,accs[i,0], c='blue')
    plt.scatter( i ,accs[i,1], c='orange')

#plt.legend(fontsize=12, loc=2)

plt.xlabel("n_value minus 5",fontsize = 12)
plt.ylabel("accuracy",fontsize = 12)

plt.xlim((0,35))
plt.ylim((.6,.9))

plt.show()
```

```python
Ctests = [.01,.1,1,10,100]
for C in Ctests:
    print ('RBF kernel model with C = ', C, ' and applied to ', 15 ,' mfcc_coefficients')
    start = time.time()
    model = SVC(kernel='rbf', C=C, max_iter=500000)
    model.fit(X_tt[:,0:30], y_tt)
    end = time.time()
    print('time training: for C = ', C, ' was ' , end - start)
    
    print('on test set')
    predicted_vector = model.predict(X_val[:,0:30])
    true_vector = y_val
    conf_mat = confusion_matrix(true_vector, predicted_vector)
    print(pd.DataFrame(conf_mat))#,
    #                 columns = ['Predicted 0', 'Predicted 1', 'Predicted 2', 'Predicted 3'],
    #                 index = ['Actual 0', 'Actual 1', 'Actual 2', 'Actual 3'])
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    print('accuracy =', accuracy)
    
    
    print('on training set')
    predicted_vector = model.predict(X_tt[:,0:30])
    true_vector = y_tt
    conf_mat = confusion_matrix(true_vector, predicted_vector)
    print(pd.DataFrame(conf_mat))#,
    #                 columns = ['Predicted 0', 'Predicted 1', 'Predicted 2', 'Predicted 3'],
    #                 index = ['Actual 0', 'Actual 1', 'Actual 2', 'Actual 3'])
    accuracy = np.diagonal(conf_mat).sum()/np.sum(conf_mat)
    print('training accuracy =', accuracy)
    print ('')
```
